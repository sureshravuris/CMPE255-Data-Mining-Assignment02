```python
from pycaret.datasets import get_data
data = get_data('DailyDelhiClimateTrain')
```

```python
import pandas as pd
import numpy as np
```

```python
# 1. Convert the 'date' column to datetime format and set it as the index
data['date'] = pd.to_datetime(data['date'])
data.set_index('date', inplace=True)

# 2. Import necessary libraries from PyCaret
from pycaret.time_series import setup, create_model, plot_model

# 3. Setup the time series environment in PyCaret
ts_setup = setup(data=data, target='meantemp', fold_strategy='sliding', fold=3, seasonal_period=365)
```

```python
model = create_model('arima')
```

```python
plot_model(model)
```

```python
plot_model(plot='train_test_split')
```

```python
from pycaret.time_series import predict_model
predict_model(model)
```

```python
from pycaret.time_series import predict_model
predict_model(model, return_pred_int=True)
```

```python
predict_model(model, alpha=0.7)
```

```python
from pycaret.time_series.forecasting.functional import compare_models
best = compare_models()
```
